# A look at energy evolution in the linear gravity waves problem over time 

```python
#cat ../diag_vel.out
```

```python
t, maxU, maxMach, ke, ge, ie, dt = np.loadtxt("../diag_vel.out", skiprows=1, unpack=True)
```

```python
# ke has been normalised already but is misssing the factor of a half 
ke = ke / 2
```

```python
#ge has not been integrated in x drn
ge = ge * 1e9 
```

```python
tot = ge + ke+ ie
dtot = tot - tot[0]
```

```python
dke = ke - ke[0]
die = ie - ie[0]
dge = ge - ge[0]
```

```python
# total energy
plt.plot(t,tot-tot[0])
```

```python
# dke
plt.plot(t,dke)
```

```python
# die
plt.plot(t,die)
```

```python
# dge 
plt.plot(t,dge)
```

```python
plt.plot(t,die - dge)
```

```python
# the total internal energy, which I think shouldn't change, also appears to oscillate
plt.plot(t,ie/ie[0]-1)
```

```python
#the kinetic energy / gravitational potential energy have the correct antiphase relation
plt.plot(t,ke/ge[0],'b', t, (ge-ge[0])/ge[0],'r')
```

```python
#the kinetic versus internal energy
plt.plot(t,dke,'g',t,die,'r',t,dge,'b')
```

```python
# however, their magnitudes do not exactly match, with an appreciable difference
plt.plot(t,ke/ge[0],'b', t, -(ge-ge[0])/ge[0],'r')
```

```python
# internal energy change and gravitational energy change are in phase - but they aren't equal?
print(ie[0]/ge[0])
plt.plot(t,ie-ie[0],'b',t,ge-ge[0],'r+')
```

```python
# mnn, magnitude of the change is quite close, at least to 1:100
print(max(abs(ie-ie[0])))
print(max(abs(ge-ge[0])))
plt.plot(t,(ie-ie[0])-(ge-ge[0]))
```
