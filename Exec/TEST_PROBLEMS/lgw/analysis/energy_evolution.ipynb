```python
import numpy as np
import matplotlib.pyplot as plt
```

# A look at energy evolution in the linear gravity waves problem over time 

```python
t_rt, maxU_rt, maxMach_rt, ke_rt, ge_rt, ie_rt, dt_rt = np.loadtxt("../rt_v_rp_old/rt.out", skiprows=1, unpack=True)
t_rp, maxU_rp, maxMach_rp, ke_rp, ge_rp, ie_rp, dt_rp = np.loadtxt("../rt_v_rp_old/rp0.out", skiprows=1, unpack=True)
```

```python
# ke has been normalised already but is misssing the factor of a half 
ke_rt = ke_rt / 2
ke_rp = ke_rp / 2
```

```python
#ge has not been integrated in x drn
ge_rt = ge_rt * 1e9 
ge_rp = ge_rp * 1e9
```

```python
tot_rt = ge_rt + ke_rt + ie_rt
tot_rp = ge_rp + ke_rp + ie_rp
```

```python
dtot_rt = tot_rt - tot_rt[0]
dke_rt = ke_rt - ke_rt[0]
die_rt = ie_rt - ie_rt[0]
dge_rt = ge_rt - ge_rt[0]
```

```python
dtot_rp = tot_rp - tot_rp[0]
dke_rp = ke_rp - ke_rp[0]
die_rp = ie_rp - ie_rp[0]
dge_rp = ge_rp - ge_rp[0]
```

```python
# total energy
plt.plot(t_rt,tot_rt,'k', t_rp, tot_rp,'b')
```

```python
plt.plot(t_rt,ke_rt,'k', t_rp, ke_rp,'b+')
```

```python
plt.plot(t_rt,dge_rt,'k', t_rp, dge_rp,'b+')
```

```python
plt.plot(t_rt,die_rt,'k', t_rp, die_rp,'b+')
```

```python
plt.plot(t_rt,die_rt,'k', t_rp, die_rp,'b')
```
