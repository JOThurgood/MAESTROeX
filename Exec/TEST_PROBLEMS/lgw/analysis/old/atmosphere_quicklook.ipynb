## Is the top and bottom of the background density actually decreasing by a factor e=2.71

```python
dy = 1e9 / 64
rho0_base = ds.point([0.,dy/2,0.])['rho0'].value[0]
rho0_top = ds.point([0.,1e9-dy/2,0.])['rho0'].value[0]

print('rho0(z=bottom cell center)={}'.format(rho0_base)) # / rho0_top / np.exp(1)
print('rho0(z=top cell center)={}'.format(rho0_top))  # / rho0_top / np.exp(1)
print('ratio={}'.format(rho0_base/rho0_top)) 
```

Think the slight descrepency from $e$ is just because this isnt z=0 and z=H, but z=dy/2 and z = H-dy/2
